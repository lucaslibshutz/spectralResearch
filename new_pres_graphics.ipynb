```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm 

def powerlaw(x,a,b):
    return a*np.power(x,b)

def gaussian(x,a,mu,sigma): #amplitude, mean, standard deviation
    return a*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
%matplotlib inline

fig,ax = plt.subplots(1,figsize=(8.1,2.39))
#plot 64 PDF functions from 0 to 4
for i in np.arange(0,128,2):
    x = np.arange(0,128,0.1)
    ax.plot(x,gaussian(x,2,i,2),color='white',linewidth=2)
ax.axis("off")
plt.savefig("C:/Users/lucas/Downloads/curves.png",bbox_inches='tight',dpi=300,transparent=True)
plt.show()
```
